```python
monthly_avg = amazon_data["Adj Close"].resample("M").mean()
monthly_avg
```

```python
monthly_avg.plot(kind="line",figsize=(13,6))
```

```python
monthly_avg.plot(kind="area",figsize=(13,6))
```

```python
delta = amazon_data["Adj Close"].diff()
delta
```

```python
avg_pc = amazon_data["Adj Close"].pct_change()
avg_pc
```

```python
rolling_avg = avg_pc.rolling(window=5).mean()
rolling_avg.plot(kind="line",figsize=(13,6))
```

```python
rolling_std = avg_pc.rolling(window=5).std()
```
