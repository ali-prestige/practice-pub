```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
polu_data = pd.read_csv("UFOPOLLUTANTS.csv")
```

```python
polu_data = polu_data.head(1000)
```

```python
len(polu_data.index)
```

```python
polu_data.head()
```

```python
sns.distplot(polu_data["NO2.AQI"])
plt.show()
```

```python
dist = sns.distplot(polu_data["NO2.AQI"],color="green")
dist.set_title("polution statistics")
dist.set_xlabel("air quality of no2")
dist.set_ylabel("frequency")
plt.show()
```

```python
rug = sns.rugplot(polu_data["NO2.AQI"],
                 height = 0.7,
                 axis = 'X')
rug.set_title("RUG PLOT")
rug.set_xlabel("air quality index")
plt.show()
```

```python
plt.subplots(figsize=(20,5))
dist = sns.distplot(polu_data['NO2.AQI'],
                   bins = 95,
                   rug = True)
plt.show()
```

```python
sns.set(color_codes = True)
plot = sns.kdeplot(polu_data["NO2.AQI"],color ="b",shade = True)
plt.show()
```

```python
joint_plot = sns.jointplot("NO2.Mean","CO.Mean",data = polu_data,kind="reg",
                          xlim ={20,60},
                          ylim={1,3})
plt.show()
```

```python
with sns.axes_style("darkgrid"):
    sns.jointplot("NO2.Mean","CO.Mean",
                 data = polu_data,
                 kind="hex",
                 )
plt.show()
```

```python
with sns.axes_style("ticks"):
    sns.jointplot("NO2.Mean","CO.Mean",
                 data = polu_data,
                 kind = "kde",
                 space = 0,
                 color = "g")
plt.show()    
```

```python
sns.kdeplot(polu_data['NO2.Mean'],
           polu_data["CO.Mean"],
           cmap = "PRGn",
           cbar = True)
plt.show()
```

```python
sns.pairplot(polu_data,
            height = 5,
            vars= ["NO2.Mean","CO.Mean","SO2.Mean","O3.Mean"])
plt.show()
```

```python
sns.pairplot(polu_data,
            height = 4,
            vars = ["NO2.Mean","CO.Mean","SO2.Mean","O3.Mean"],
            hue = "year")
plt.show()
```

```python
sns.pairplot(polu_data,
            height = 4,
            vars = ["NO2.Mean","CO.Mean","SO2.Mean","O3.Mean"],
            kind = "reg")
plt.show()
```
